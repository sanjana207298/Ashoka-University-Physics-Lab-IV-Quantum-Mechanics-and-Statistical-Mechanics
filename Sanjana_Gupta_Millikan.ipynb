```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['font.family'] = 'serif'
plt.rcParams['font.serif'] = ['Times New Roman'] + plt.rcParams['font.serif']
```

```python
#Constants

b = 8.20e-3
p = 1.01e5
g = 9.81
eta = 1.844e-5
rho_air = 1.184
rho_drop = 895
d = 0.01
l = 1e-3
e = 1.602176634e-19
```

## Without Terminal Velocity

```python
#Drop 1a
V = 0.26*1000
#rising
t2 = np.array([17.35,14.5,20.49,21.23,26.81])
v2 = l/t2
#falling
t1 = np.array([14.42,13.91,14.66,14.25,13.16])
v1 = l/t1

r1a = (np.sqrt(((b**2)/(4*(p**2))) + ((9*eta*v1)/(2*g*rho_drop))) - (b/(2*p)))
q1a = ((v2/v1) + 1)*(4*np.pi*(r1a**3)*d*(rho_drop-rho_air)*g)/(3*V)

print('Radius of the drop is:', r1a , 'm')
print('Total Charge in the drop =', q1a , 'Coulombs')
print('Number of electrons in the drop =', (q1a/e))
print(np.std(v1))
print(np.std(v2))
err = (1/V)**2 + 5/2*(np.std(v1)/len(v1))**2 + (np.std(v2)/len(v2))**2
print(np.sqrt(err)*np.mean(q1a))
```
